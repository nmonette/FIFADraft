```python
import pandas as pd
```

```python
data = pd.read_csv("fifa23.csv")
data["id"] = [i for i in range(data.shape[0])]
data.columns
```

```python
data = data.drop(columns=["Image Link", "Wage(in Euro)", 'Release Clause', 'Contract Until', 'Club Jersey Number', 'Joined On', 'On Loan', 'International Reputation', 'National Team Image Link', 'National Team Jersey Number'])

data = data[["id"] + list(data.columns)[:-1]]
```

```python
data.to_json("fifa23.json", orient="records")
```
